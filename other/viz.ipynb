```python
IN = "../data/cvrp_670.json"
```

```python
import sys
sys.path.insert(0, "..")
import json
import matplotlib.pyplot as plt
from CVRPSolver import solve

with open(IN) as f:
    instance = json.load(f)

locs = instance["Coordinates"]
demands = instance["LocationDemands"]
num_of_vehicles = instance["NumberOfVehicles"]
capacity = instance["VehicleCapacity"]
dist_mat = instance["DistanceMatrix"]
sol = solve(dist_mat, locs, demands, num_of_vehicles, capacity)
```

```python
plt.scatter([x for x, _ in locs], [y for _, y in locs])
plt.show()

for route in sol:
    route = [0] + route + [0]
    plt.plot([locs[loc][0] for loc in route], [locs[loc][1] for loc in route])
plt.show()

for route in instance["GlobalBestSolution"]:
    plt.plot([locs[loc][0] for loc in route], [locs[loc][1] for loc in route])
plt.show()
```
