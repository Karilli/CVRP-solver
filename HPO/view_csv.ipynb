```python
import pandas as pd
```

```python
df = pd.read_csv('./SMAC-small.csv')
df = df.loc[df.groupby('summary:config_id')['summary:budget'].idxmax()].reset_index(drop=True)
X = [
    (15, "metric:score (minimize)"),
    (4, "summary:cvrp_66"),
    (4, "summary:cvrp_78"),
    (4, "summary:cvrp_80")
]
CONFIGS = [
    ({col[12:]: df[col][row] for col in df.columns if "CVRP" in col and "ALLOW" not in col}, df["metric:score (minimize)"][row])
    for n_configs, col_to_minimize in X
    for row in df.nsmallest(n_configs, col_to_minimize).index
]

UNIQUE = {str(dct): (i, score) for i, (dct, score) in enumerate(CONFIGS[::-1])}
[dct[0] for _, dct in sorted([(score, CONFIGS[idx]) for idx, score in UNIQUE.values()])]
```

```python
df.nsmallest(10, "metric:score (minimize)")[["metric:score (minimize)", "summary:cvrp_32", "summary:cvrp_45", "summary:cvrp_55", "summary:cvrp_66", "summary:cvrp_78", "summary:cvrp_80"]]
```
